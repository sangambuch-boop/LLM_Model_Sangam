```python
import pandas,numpy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pandas.read_csv("D:\\irish_data\\Boston-house-price-data.csv")
x = df.drop(columns=["MEDV"])
y = df[["MEDV"]]
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
# ypred = model.predict(X_test)
```

```python
# mes = mean_squared_error(y_test,ypred)
# r2 = r2_score(y_test,ypred)

# Ab accuracy check karein
ypred = model.predict(X_test)
r2 = r2_score(y_test, ypred)
print(f"Nayi Accuracy (All Features): {r2 * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn
```

```python
plt.figure(figsize=(8,6))

plt.scatter(y_test,ypred,color="blue",label="Ghron Ka data",alpha=0.6)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color="red",linestyle="--",label="perfect predicition Line",lw=2)


# Labels aur Title lagana

plt.title('Asli Keemat vs Model ki Predicted Keemat (Boston Dataset)', fontsize=14)
plt.xlabel('Asli Keemat (Actual Price - MEDV)', fontsize=12)
plt.ylabel('Model ki Batayi Keemat (Predicted Price)', fontsize=12)
plt.legend()

plt.grid(True)

plt.show()
```
